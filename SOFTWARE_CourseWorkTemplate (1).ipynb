```python
#import the numpy library
import numpy as np
#filling up the missing values in each column using sklearn simple imputer 
from sklearn.impute import SimpleImputer
#create a function to fill missing values using simple imputers
def mean_median(df,column):
    #addressing the numerical features
    if type(df[column]) is ['int64','float64']:
        imputer=SimpleImputer(strategy='mean')
        #fit the imputer into the column
        imputer.fit(df[[column]])
        df[[column]]=imputer.transform(df[[column]])
    else:
        #addressing the categorical features
        cat_imputer=SimpleImputer(strategy='most_frequent')
        #fit the imputer into the column
        cat_imputer.fit(df[[column]])
        df[[column]]=cat_imputer.transform(df[[column]])
        
    def mean(values):
        mean=sum(values)/len(values)
        return mean
    return df.head()
mean_median(raw_data,'Calories')
```
